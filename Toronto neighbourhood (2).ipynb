```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
print('Data downloaded!')
```

```python
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "lxml")
all_tables=soup.find("table")
all_tables
```

```python

postalcode=[]
borough=[]
neighbourhood=[]
for row in all_tables.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        postalcode.append(cells[0].find(text=True))
        borough.append(cells[1].find(text=True))
        neighbourhood.append(cells[2].find(text=True))
```

```python
df=pd.DataFrame(postalcode,columns=['Postal Code'])
df['borough']=borough
df['neighbourhood']=neighbourhood
df.head()
```

```python
df1=df[df.borough!='Not assigned\n']
df1
```

```python
df1.loc[df1['neighbourhood']=='Not assigned\n', 'neighbourhood']=df1['borough']
df1 = df1.replace('\n','', regex=True)
df1
```

```python
df1.shape
```

```python
df2=pd.read_csv("http://cocl.us/Geospatial_data")
df2
```

```python
df3=pd.merge(df1,df2, how='outer', on=['Postal Code'])
df3
```

```python
toronto_grouped=df3.loc[df3['borough'].str.contains("Toronto")]
toronto_grouped
```
